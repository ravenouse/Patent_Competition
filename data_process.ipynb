To include the title information

```python
train_df = pd.read_csv("./datasets/train.csv")
titles = pd.read_csv("./datasets/titles.csv")
```

```python
train_df = train_df.merge(titles, left_on='context', right_on='code')
```

```python
train_df['input'] = "titles:"+train_df['title']+' '+train_df['anchor']
```

```python
train_df.to_csv('./datasets/train_titles.csv')
```
